**Задача 1.**

Для начала заметим, что $r \cdot x_{n + 1} = x_n (1 - x_n) = r (x_n - x^2_n).$ Исследуя параболу $x_n - x^2_n$ (то есть понимая, что у неё два корня: $0$ и $1$, а ордината вершины, находящейся между ними равна $0.5 - 0.5^2 = 0.25$) получим что на отрезке $[0; 1]$ она принимает значения из отрезка $[0; \frac14].$ Получается, если коэффициент $r$ будет из отрезка $[0; 1],$ то $x_{n + 1} \in [0; \frac14].$ Учитывая, что $x_0 \in [0; 1]$ то индуктивно каждый следущий точно будет из отрезка $[0; \frac14].$ Получили более строгую оценку (начиная со второго члена последовательности).

**Задача 2.**

Построим графики зависимости значения члена $x_n$ от его номера.

При $r \in [0; 4]:$ функция монотонно убывает в положительных значениях.


При $r > 4$ функция имеет несколько первых положительных значения, а далее резко убывает к $ - ∞.$

При $r < 0$ функция становится знакопеременной, при этом убывает по модулю.

(Посмотреть примеры можно, помеяв значение параметра $r$ в функции *logistic_function*)

```python
import matplotlib.pyplot as plt

def logistic_function(x_n, r=-1):
    return r * x_n * (1 - x_n)

x = [i for i in range(1, 101)]

y = [0.5]
while len(y) != 100:
    y.append(logistic_function(y[-1]))

plt.plot(x, y)
plt.show()
```

А именно, функция $x_{n + 1} = r \cdot x_n(1 - x_n)(2 - x_n)$ при $r \in [0; \frac{3\sqrt3}2].$

Заметим, что ордината локального максимума функции $x_n(1 - x_n)(2 - x_n)$ есть $\frac{2\sqrt3}9,$ при том $f_{max} \cdot r_{max} = \frac{2\sqrt3}9 \cdot \frac{3\sqrt3}2 = 1.$ Это означает, что при данных $r$ значения функции не выйдут за пределы $[0; 1].$

```python
from math import sqrt

def function(x_n, r=-1):
    return r * x_n * (1 - x_n) * (2 - x_n)

x = [i for i in range(1, 101)]

y = [0.5]
while len(y) != 100:
    y.append(function(y[-1]))

plt.plot(x, y)
plt.show()
```
