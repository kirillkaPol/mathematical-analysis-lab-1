```python
''' Задача 1. Вспомогательные функции. '''
def rev(a: int) -> int:
  a = str(a)
  a = a[::-1]
  a = int(a)
  return a

def is_polyndrom(a: int) -> bool:
  return a == rev(a)

def is_prime(a: int) -> bool:
  if a < 2:
    return False
  edge = int(sqrt(a)) + 1
  for i in range(2, edge):
    if a % i == 0:
      return False
  return True

def circular_changes(a: int) -> list:
  a = list(str(a))
  changes = []
  for i in range(len(a)):
    popped = a.pop(0)
    a.append(popped)
    changes.append(int("".join(a)))
  return changes
```

```python
''' Задача 1. Реализация. '''
def palindromic_squares_and_circular_primes() -> tuple[List[int], List[int]]:
  task1: List[int] = []
  for a in range(1, 10**5):
    if is_polyndrom(a):
      if is_polyndrom(a**2):
        task1.append(a)

  task2: List[int] = []
  for p in range(1, 10**6):
    if is_prime(p):
      if all(map(is_prime, circular_changes(p))):
        task2.append(p)

  return task1, task2

palindromic_squares, circular_primes = palindromic_squares_and_circular_primes()
print("Палиндромические квадраты:", *palindromic_squares[:10], "...")
print("Циклические простые:", *circular_primes[:10], "...")
```

```python
''' Задача 3. Вспомогательная функция. '''
def to_new_system(num, base):
  if num == 0:
    return "0"
  new = ''
  while num > 0:
      remainder = num % base
      new = str(remainder) + new
      num = num // base

  return new

def make_number(a: int, b: int, ind: int) -> tuple[int, int]:
  ind = to_new_system(ind, 3)
  if "0" in ind:
    return 4
  d = [None, min(a,b), max(a,b)]
  num = ""
  for i in ind:
    num += str(d[int(i)])
  return int(num)

def primes_with_digits(a: int, b: int) -> List[int]:
      result = []
      i = 0
      while len(result) <= 100:
        num = make_number(a, b, i)
        if is_prime(num):
          result.append(num)
        i += 1
      return result
```

```python
''' Задача 6. Реализация. '''
def euler_phi_direct(n: int) -> int:
    count = 0
    for k in range(1, n + 1):
        if gcd(k, n) == 1:
            count += 1
    return count

def euler_phi_factor(n: int) -> int:
    factors = factorint(n)
    phi = n
    for p in factors:
        phi *= (1 - 1 / p)
    return int(phi)

def compare_euler_phi_methods(test_values: List[int]) -> dict[str, List[float]]:
    times_direct, times_factor, times_sympy = [], [], []
    for n in test_values:
        start_time = time.time()
        euler_phi_direct(n)
        times_direct.append(time.time() - start_time)

        start_time = time.time()
        euler_phi_factor(n)
        times_factor.append(time.time() - start_time)

        start_time = time.time()
        totient(n)
        times_sympy.append(time.time() - start_time)

    return {'direct': times_direct,
            'factor': times_factor,
            'sympy': times_sympy}

test_values = [10, 50, 100, 500, 1000, 5000, 10000]
times = compare_euler_phi_methods(test_values)
for i, n in enumerate(test_values):
    print(f"n = {n:5d} | direct = {times['direct'][i]:.6f}s | "
          f"factor = {times['factor'][i]:.6f}s | sympy = {times['sympy'][i]:.6f}s")
```

```python
''' Вспомогательные функции и классы. '''
def permToCycle(perm_set: list) -> list:
    result = []
    used = []
    for i in range(1, len(perm_set) + 1):
        if i not in used:
            used.append(i)
            cycle = [i]
            while True:
                next = perm_set[cycle[-1] - 1]
                used.append(next)
                cycle.append(next)
                if len(cycle) != len(set(cycle)):
                    cycle.pop(-1)
                    break
            result.append(cycle)
    return(result)

class Permutation:
    def __init__(self, perm_set: list):
        self.perm_set = perm_set
        self.degree = len(perm_set)
    
    def __eq__(self, other):
        if isinstance(other, Permutation):
            return self.perm_set == other.perm_set
        return False

    def __repr__(self):
        perm_cycle = permToCycle(self.perm_set)
        output = []
        for cycle in perm_cycle:
            cycle[0] = f"({cycle[0]}"
            cycle[-1] = f"{cycle[-1]})"
            for i in range(len(cycle)):
                cycle[i] = str(cycle[i])
            cycle = " ".join(cycle)
            output.append(cycle)

        return "".join(output)

    def __mul__(self, other):
        result = [0 for i in range(self.degree)]
        for i in range(self.degree):
            result[i] = other.perm_set[self.perm_set[i] - 1]
        return Permutation(result)

    def __pow__(self, power: int):
        if power == 0:
            return Permutation([i for i in range(1, len(self) + 1)])
        if power == 1:
            return self
        if power > 1:
            return self * self.__pow__(power - 1)

    def __len__(self):
        return len(self.perm_set)


def lcm(*args):
    if len(args) == 1:
        return args[0]
    if len(args) == 2:
        return abs(args[0] * args[1]) // gcd(args[0], args[1])
    else:
        return lcm(args[-1], lcm(*tuple(list(args)[:-1])))

from itertools import permutations

class Permutation_group:
    def __init__(self, m):
        self.m = m
        pset = [i for i in range(1, m + 1)]
        perms = list(permutations(pset))
        self.elems = [Permutation(list(i)) for i in perms]


    def __repr__(self):
        return 'S' + str(self.m)

    def __len__(self):
        return factorial(self.m)


class Group_set:
    def __init__(self, group_set: list, group_ord: int):
        self._group_set = group_set
        self._group_ord = group_ord

    def __getitem__(self, index):
        return self._group_set[index % self._group_ord]
    # не делать цыклы for i in group_set (он будт бесконечным)

    def __setitem__(self, index, value):
        self._group_set[index] = value

    def __len__(self):
        return len(self._group_set)

class Cyclic_permutation_group:
    def __init__(self, perm: Permutation):
        self.gen_perm = perm
        self.group_ord = lcm(*[len(i) for i in permToCycle(perm.perm_set)])
        self.group_set = Group_set([0 for i in range(self.group_ord)], self.group_ord)
        for power in range(1, self.group_ord + 1):
            if power == self.group_ord:
                self.group_set[0] = perm ** power
            else:
                self.group_set[power] = perm ** power
        self.group_neutral = self.group_set[0]

def is_group(set, operation):
  # проверка, являеься, ли закон внутренним
  for elem1 in set:
      for elem2 in set:
          if operation(elem1, elem2) not in set:
              return False

  # операция - умножение по модулю. Она, очевидно, ассоциативна.

  # проыерка на наличие нейтрального
  flag = False
  for n_elem in set:
      for elem in set:
          if operation(n_elem, elem) != elem:
              break
      else:
          neutral = n_elem
          flag = True
          break
  if not flag:
      return False

  # проверка на наличие обратных
  for elem1 in set:
      flag = False
      for elem2 in set:
          if operation(elem1, elem2) == neutral:
              flag = True
      if not flag:
          return False

  return True

def subsets(elements):
  n = len(elements)
  subsets = []
  for mask in range(1 << n):
      subset = []
      for i in range(n):
          if mask & (1 << i):
              subset.append(elements[i])
      subsets.append(subset)

  return subsets
```

```python
'''  Задача 1 '''
def subgroups_of_Sm(N: int) -> dict:
    m = 3 
    Sm = Permutation_group(m)
    elements = Sm.elems
    n = len(elements)
    all_subgroups = []

    for mask in range(1, 1 << n):
        subset = []
        for i in range(n):
            if (mask >> i) & 1:
                subset.append(elements[i])
        
        if is_group(subset, lambda x,y: x*y):
            all_subgroups.append(subset)

    num_subgroups = len(all_subgroups)
    random_subgroup = random.choice(all_subgroups)

    target_idx = N % num_subgroups
    H = all_subgroups[target_idx]
    
    left_cosets = set()
    right_cosets = set()
    
    for g in elements:
        l_coset = tuple(sorted([tuple((g * h).perm_set) for h in H]))
        r_coset = tuple(sorted([tuple((h * g).perm_set) for h in H]))
        left_cosets.add(l_coset)
        right_cosets.add(r_coset)
    
    index_H = len(left_cosets)
    is_normal = (left_cosets == right_cosets)

    return {
        "count": num_subgroups,
        "random_subgroup": random_subgroup,
        "target_subgroup": H,
        "index": index_H,
        "is_normal": is_normal,
        "left_cosets": left_cosets,
        "right_cosets": right_cosets
    }

print(subgroups_of_Sm(N))
```

```python
''' Задача 6. Вспомогательная функции. '''
def gpow(elem, power, func):
    if power == 1:
        return elem
    if power > 1:
        return func(elem, gpow(elem, power - 1, func))

def order(elem, g_ord):
    pow = 0
    previois = []
    while True:
        pow += 1
        result = (elem ** pow) % g_ord
        if result in previois:
            return float('inf')
        if result == 1:
            return pow
        previois.append(result)
```

```python
''' Задача 11. Реализация. '''
def find_polynomial_roots_with_coefficients(N):
    x = symbols('x')
    # Вычисляем коэффициенты
    a_coeffs = [(i + N) % 4 for i in range(9)]
    b_coeffs = [(j + N) % 7 for j in range(7)]

    roots_F4 = Poly(x ** 9 + sum(a_coeffs[i] * x ** i for i in range(9)), x, domain=GF(4)).nroots()

    roots_F7 = Poly(sum(b_coeffs[i] * x ** i for i in range(7)), x, domain=GF(7)).nroots()

    return {
        'N': N,
        'a_coeffs': a_coeffs,
        'b_coeffs': b_coeffs,
        'roots_F4': roots_F4,
        'roots_F7': roots_F7
    }

result = find_polynomial_roots_with_coefficients(N)
for key, word in result.items():
    print(key, word)
```

```python
''' Задача 13. Реализация. '''
def gcd_and_linear_representation(N: int) -> dict:
    r_coeffs = [(m + N) % 11 for m in range(8)]
    s_coeffs = [(t + N) % 11 for t in range(4)]

    x = symbols('x')
    f = Poly(sum(r_coeffs[i] * x**i for i in range(8)), x, domain=GF(11))
    g = Poly(sum(s_coeffs[i] * x**i for i in range(4)), x, domain=GF(11))

    f, g, d = gcdex(f, g)
    return {'gcd': d, 'f': f, 'g': g}

task13 = gcd_and_linear_representation(N)
print("НОД полиномов над GF(11):", task13['gcd'])
print("f(x):", task13['f'])
print("g(x):", task13['g'])
print(f'gcd(f, g) = {task13['gcd']} * {task13['f']} + {task13['gcd']} * {task13['g']}')
```
